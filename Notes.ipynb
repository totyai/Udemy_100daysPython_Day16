Object Oriented Programing

- Procedual Programing is when the program goes liners, going from here to there inwith fucntions
- Object Oriented Programing is solving and abstracting the things, so it can be resuseable for different components even if the end product is different, somethings can be used
- A benefit for OOP is also that multiple teams can work on hte same project simultaniously, on different moduls
- In OOP, everything is an object, and would represent real world objects AN object is a collection of *attributes* and *methods* (variables and fucntions), that define them.
- These objects are definitions, but there can be multiple instances of an object *blueprint*. The blueprint is stored in *class*es and the indivudual instaces *object*.
- In Python the Class is defined with each word starting with captial case and together.
- We can create an intace of a class by assigning it as variable:
```
car = CarBlueprint()
```
- Every object has attributes, informations that it stores
- Accessing them is used by the variable name . attibute
```
car.speed
```
- Every object has methods, these are telling what does the object do
- Accessing them us used by the variable name . method
```
car.stop()
```
